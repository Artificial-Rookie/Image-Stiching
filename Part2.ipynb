## Part One 2D Convolution

```python
import cv2 
import matplotlib.pyplot as plt

# reading the image
img = cv2.imread('im01.jpg')
# create SIFT feature extractor
sift = cv2.SIFT_create()
# detect features from the image
keypoints, descriptors = sift.detectAndCompute(img, None)
# draw the detected key points
sift_image = cv2.drawKeypoints(img, keypoints, outImage=None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
# show the image
sift_image = cv2.cvtColor(sift_image, cv2.COLOR_BGR2RGB)
plt.imshow(sift_image)
plt.title("sift features")
plt.show()
```
